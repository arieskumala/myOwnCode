## Enkripsi dengan metode Caesar:
Metode merupakan metode yang paling sederhana dalam melakukan enkripsi. Teknik ini hanya melakukan pemetaan suatu karakter huruf (angka dan karakter spesial tidak termasuk) dengan metode pergeseran karakter. contoh

| huruf | shift 2 | shift 3 | shift 4 | shift -2 |
|-------|---------|---------|---------|----------|
| a     | c       | d       | e       | y        |
| A     | C       | D       | E       | Y        |
| c     | e       | f       | g       | a        |
| z     | b       | c       | d       | x        |
| %     | %       | %       | %       | %        |
| _     | _       | _       | _       | _        |
| 5     | 5       | 5       | 5       | 5        |

ketentuan:
* pergeseran satu huruf setelah `z` adalah `a`.
* Jika input karakter merupakan huruf kapital maka output juga kapital, begitu juga dengan huruf kecil/non-kapital.
* semua yang bukan huruf tidak berubah, jadi kalau ada suatu angka maka karakter enkripsinya tetap angka tersebut, begitu juga dengan semua karater spesial termasuk spasi.

## Langkah Penyelesaian

Saya akan menggunakan fungsi `ord`, `chr` dan string method `.isalpha()` dan `.isupper`untuk menyelesaikan soalnya ini.

Ok, langkah pertama yaitu saya akan membuat perulangan sebanyak jumlah karakter dalam pesan `txt` selanjutnya dilakukan pengecekkan apakah karakter tersebut adalah alfabet (a,b,c..,z) atau bukan, menggunakan conditional statement `if ... else ...`. Jika ya, maka lakukan shifting sesuai dengan ketentuan dan cetak hasilnya kedalam variabel `ciphertext`. Jika bukan, maka tidak perlu dilakukan shifting, jadi lansung dicetak saja karakternya kedalam variabel `ciphertext`. Dan kembalikan hasilnya dengan menggunakan `return`.

````
def caesar_encript(txt,shift):
  ciphertext=''
  for char in txt:
    if char.isalpha() == True:
      if char.isupper() == True: 
        encrypt = chr((((ord(char) + shift) - 65) % 26) + 65)   
        ciphertext += encrypt
      else:
        encrypt = chr((((ord(char) + shift) - 97) % 26) + 97)
        ciphertext += encrypt
    else: 
      ciphertext += char
  return ciphertext 
    
# Fungsi Decript caesar
def caesar_decript(chiper,shift):
  return caesar_encript(chiper,-shift)
````

````
# Sanity check!!!

msg = 'Haloz DTS mania, MANTAPSZZZ!'
cpr = caesar_encript(msg,4)
txt = caesar_decript(cpr,4)

print('plain text:',txt)
print('chiper text:',cpr)
````

### Expected output:
```
plain text: Haloz DTS mania, MANTAPSZZZ!
chiper text: Lepsd HXW qerme, QERXETWDDD!
```

Buatlah fungsi `deshuffle_order`, dengan argument `sftxt` (string) dan `order` (list). Yang akan mengembalikan urutan string kembali seperti semula sebelum di-shuffle. Sedemikian hingga: 
`deshuffle_order(shuffle_order(txt,order),order) == txt`
 
**hint: list method `.index()`**

## Langkah Penyelesaian
List order yang diberikan merepresentasikan posisi huruf diacak dengan ketentuan tersebut. Misal huruf `a` dengan index `0` posisinya akan diubah menjadi diurutan ke-4, karena didalam list order angka `0` ada dipaling belakang. Huruf `b` dengan index `1` posisinya akan diubah menjadi diurutan ke-dua, kebetulan ini sama, dan begitu seterusnya.

Nah artinya untuk mengembalikan urutannya, kita bisa melakukan indexing terhadap `order = [2, 1, 3, 0]` untuk menemukan dimana posisi angka 0 sampai 3 berada. Misal `print(order.index(0))` outputnya adalah **3**, karena angka 0 berada diindex **ke-3** dari list order, dan seterusnya, lebih lanjut bisa dibaca [disini](https://www.w3schools.com/python/ref_list_index.asp)

Kita manfaatkan method `.index` ini untuk mengurutkan huruf di `sftxt`. Dengan menggunakan syntax 

`sftxt[order.index(0)]` = `sftxt[3]` = a

`sftxt[order.index(1)]` = `sftxt[1]` = b

`sftxt[order.index(2)]` = `sftxt[0]` = c

`sftxt[order.index(3)]` = `sftxt[2]` = d

untuk memudahkan operasinya tinggal dilakukan perulangan saja sebanyak panjang ordernya. Siapkan sebuah variabel `result` untuk menampung hasil perulangan. Dan kembalikannya hasilnya dengan menggunakan `return`.

````
def shuffle_order(txt,order):
  return ''.join([txt[i] for i in order])
 
def deshuffle_order(sftxt,order):
  result = ""
  for i in range(len(order)):
    result += sftxt[order.index(i)] 
  return result 
````

````
# Sanity check!!!
 
print(shuffle_order('abcd',[2,1,3,0]))
print(deshuffle_order('cbda',[2,1,3,0]))
````

### Expected output:
```
cbda
abcd
```

Buatlah fungsi `send_batch` dengan argument `txt` (string), `batch_order` (list), dan `shift` (int). Fungsi ini akan memecah text menjadi bagian-bagian yang lebih kecil sesuai dengan panjang dari `batch_order`, di mana batch_order tersebut merupakan list index untuk men-shuffle setiap batch dengan menggunakan fungsi di nomor 2. Keluaran fungsi ini berupa list batch teks terenkripsi.   
 
**Note: tambahkan underscore di akhir (sebelum dipecah menjadi batches) jika panjang `txt` belum memenuhi kelipatan panjang satu batch yaitu `len(batch_order)`.**
 
**Hint: Lihat intro 3 dan gambar, gunakan library `math` jika diperlukan**

##Langkah Pengerjaan
Lakukan enkripsi terhadap pesan (txt) dengan menggunakan fungsi nomor 1. Karena pesan akan kita menjadi sepanjang `batch_order`, cek apakah pesan tersebut habis dibagi dengan panjang `batch_order`, jika tidak tambahkan underscore dengan menggunakan while loop statement.

Jika sudah baru kemudian pesan (txt) dibagi menjadi sepanjang `batch_order`, dan lakukan pengacakkan dengan menggunakan fungsi shuffle pada soal nomor 2.

kembalikannya hasilnya dengan menggunakan `return`

````
def send_batch(txt, batch_order, shift):
    ciphertext = caesar_encript(txt, shift) 
    while len(ciphertext) % len(batch_order) != 0:
        ciphertext += "_"                                             
    ciphertext = ([ciphertext[i:i + len(batch_order)] for i in range(0, len(ciphertext), len(batch_order))])
    return ([shuffle_order(i, batch_order) for i in ciphertext])

def receive_batch(batch_cpr, batch_order, shift=3):
    batch_txt = [caesar_decript(deshuffle_order(i, batch_order), shift) for i in batch_cpr]
    return ''.join(batch_txt).strip('_')
````

````
# Sanity check!!!
 
msg_cpr = send_batch('halo DTS mania, mantaaap!!!',[2,1,3,0],4)
msg_txt = receive_batch(msg_cpr,[2,1,3,0],4)
print(msg_txt,msg_cpr,sep='\n')
````
